```python
import numpy as np
from tqdm import tqdm
import torch
```

```python
emb = np.load("../IllinoisGraphDataset_v1/paper_384_emb_1.npy", allow_pickle=True)
```

```python
paper_ids_emb = []
paper_ids = []
for paper_id, paper_emb in emb.item().items():
    paper_ids_emb.append(paper_emb)
    paper_ids.append(paper_id)
```

```python
len(paper_ids_emb)
```

# Create mapping

```python
paper_idx_to_emb = {}
for i, element in enumerate(paper_ids):
    paper_idx_to_emb[element] = i+1
```

```python
max(paper_idx_to_emb.values()) == len(paper_ids_emb)
```

```python
required_edges = []
```

```python
paper_ids_set = set(paper_ids)
```

```python
edges = np.load("../IllinoisGraphDataset_v1/cites_edge_1.npy", mmap_mode='r')
```

```python
for edge in tqdm(edges):
    if edge[0] in paper_ids_set and edge[1] in paper_ids_set:
        required_edges.append([paper_idx_to_emb[edge[0]], paper_idx_to_emb[edge[1]]])
```

```python
len(required_edges)
```

## Labels

```python
labels = np.load("../IllinoisGraphDataset_v1/paper_ss_19_labels.npy", allow_pickle=True)
```

```python
paper_id_labels = {}
for paper_id, label in tqdm(labels.item().items()):
    if paper_id in paper_ids_set:
        paper_id_labels[paper_idx_to_emb[paper_id]] = label
```

```python
len(paper_id_labels)/len(paper_ids_emb)
```

```python
paper_id_labels = dict(sorted(paper_id_labels.items()))
```
